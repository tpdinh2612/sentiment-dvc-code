## 4. Switch between versions

```python
# Thay mã hash commit
hash_commit = "76527e6"
!git checkout {hash_commit} data/raw/sentiment.csv.dvc
```

Vì:
- Git chỉ quản lý và lưu trữ metadata (các file code, file .dvc). Git hoàn toàn không chứa dữ liệu lớn thực tế.

- DVC lưu trữ dữ liệu lớn thực tế (dataset, file model .pt, .h5, ...) bên trong thư mục bộ nhớ đệm cục bộ (.dvc/cache) hoặc trên máy chủ từ xa (Remote Storage như S3, Google Drive)

Nên sau khi git checkout version cũ thành công, git chỉ đổi file .dvc, cần chạy `dvc pull` để tải dữ liệu từ Remote Storage.
